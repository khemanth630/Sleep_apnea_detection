```python
import os
import wfdb
import numpy as np
import pandas as pd
from scipy.signal import butter, filtfilt, welch
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Paths
data_path = '/content/drive/MyDrive/apnea-ecg/'
output_csv = '/content/drive/MyDrive/apnea-ecg/features.csv'

# Filter function
def bandpass_filter(signal, fs, lowcut=0.5, highcut=45.0):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(4, [low, high], btype='band')
    return filtfilt(b, a, signal)

# Normalize
def normalize(signal):
    return (signal - np.mean(signal)) / np.std(signal)

# Feature extraction
def extract_features(segment, fs):
    # Time domain features
    mean_val = np.mean(segment)
    std_val = np.std(segment)
    max_val = np.max(segment)
    min_val = np.min(segment)
    range_val = max_val - min_val

    # Frequency domain - Power Spectral Density
    freqs, psd = welch(segment, fs=fs)
    psd_mean = np.mean(psd)
    psd_max = np.max(psd)
    psd_power = np.sum(psd)

    return {
        'mean': mean_val,
        'std': std_val,
        'max': max_val,
        'min': min_val,
        'range': range_val,
        'psd_mean': psd_mean,
        'psd_max': psd_max,
        'psd_total_power': psd_power
    }

# Load + process
record_list = ['a01', 'a02', 'b01']
samples_per_minute = 60

all_feature_rows = []

for record_name in record_list:
    try:
        # Load ECG signal
        record = wfdb.rdrecord(os.path.join(data_path, record_name))
        ecg_signal = record.p_signal[:, 0]
        fs = int(record.fs)

        ecg_signal = bandpass_filter(ecg_signal, fs)
        ecg_signal = normalize(ecg_signal)

        # Load apnea annotations
        ann = wfdb.rdann(os.path.join(data_path, record_name), 'apn')
        positions = ann.sample
        labels = ann.symbol

        for i, start_sample in enumerate(positions):
            end_sample = start_sample + fs * samples_per_minute
            if end_sample > len(ecg_signal):
                continue

            segment = ecg_signal[start_sample:end_sample]
            label = labels[i]

            features = extract_features(segment, fs)
            features['label'] = label
            all_feature_rows.append(features)

            print(f"Segment {i+1} | Label: {label} | Features: {features}")

    except Exception as e:
        print(f"Error processing record {record_name}: {e}")

# Save to CSV
df = pd.DataFrame(all_feature_rows)
df.to_csv(output_csv, index=False)
print(f"\n✅ Features saved to {output_csv}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.utils import to_categorical

# ============================
# 1. Load Features Data (features.csv)
# ============================
features_df = pd.read_csv('/content/drive/MyDrive/apnea-ecg/processed_data/features.csv')

# Print to check the first few rows
print(features_df.head())

# ============================
# 2. Prepare Data
# ============================
# Assume 'Label' is the target column and the rest are features
X = features_df.drop('Label', axis=1).values
y = features_df['Label'].values

# Encode the labels into numbers
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)

# Convert labels to one-hot encoding
y_categorical = to_categorical(y_encoded)

# ============================
# 3. Train-Test Split
# ============================
X_train, X_test, y_train, y_test = train_test_split(X, y_categorical, test_size=0.2, random_state=42)

# Reshape for LSTM: (samples, time_steps, features)
# LSTM expects input in the form of (samples, timesteps, features), so reshape the data
X_train = X_train.reshape(X_train.shape[0], 1, X_train.shape[1])  # Reshaping to (samples, 1, features)
X_test = X_test.reshape(X_test.shape[0], 1, X_test.shape[1])  # Reshaping to (samples, 1, features)

# ============================
# 4. LSTM Model
# ============================
lstm_model = Sequential([
    LSTM(64, return_sequences=False, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(2, activation='softmax')  # Output layer with 2 classes (e.g., 'A' and 'N')
])

# Compile the model
lstm_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Print model summary
lstm_model.summary()

# ============================
# 5. Train the Model
# ============================
lstm_model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))

# ============================
# 6. Evaluate the Model
# ============================
lstm_eval = lstm_model.evaluate(X_test, y_test, verbose=0)
print(f"✅ LSTM Model Accuracy: {lstm_eval[1]*100:.2f}%")

# ============================
# 7. Predict and Print Results
# ============================
y_pred = lstm_model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)  # Convert predictions to class labels

# Convert one-hot encoded true labels back to class labels
y_test_classes = np.argmax(y_test, axis=1)

# Calculate accuracy
accuracy = accuracy_score(y_test_classes, y_pred_classes)
print(f"Accuracy on test data: {accuracy * 100:.2f}%")
```

```python
import os
import wfdb
import numpy as np
import pandas as pd
from scipy.signal import butter, filtfilt, find_peaks
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Path settings
data_path = '/content/drive/MyDrive/apnea-ecg/'
output_csv = '/content/drive/MyDrive/apnea-ecg/rr_features.csv'

# Bandpass Filter
def bandpass_filter(signal, fs, lowcut=0.5, highcut=45.0):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(4, [low, high], btype='band')
    return filtfilt(b, a, signal)

# Normalize
def normalize(signal):
    return (signal - np.mean(signal)) / np.std(signal)

# Extract RR features from a 1-minute ECG segment
def extract_rr_features(ecg_segment, fs):
    peaks, _ = find_peaks(ecg_segment, distance=fs*0.6)  # Minimum 0.6s between peaks
    rr_intervals = np.diff(peaks) / fs  # Convert samples to seconds

    if len(rr_intervals) == 0:
        return {f"rr_{stat}": 0 for stat in ['mean', 'std', 'min', 'max', 'count']}

    features = {
        'rr_mean': np.mean(rr_intervals),
        'rr_std': np.std(rr_intervals),
        'rr_min': np.min(rr_intervals),
        'rr_max': np.max(rr_intervals),
        'rr_count': len(rr_intervals)
    }
    return features

# Records list
record_list = [f"a{str(i).zfill(2)}" for i in range(1, 21)] + \
              [f"b{str(i).zfill(2)}" for i in range(1, 6)] + \
              [f"c{str(i).zfill(2)}" for i in range(1, 11)]

# Loop through each record
features_data = []

for record in record_list:
    try:
        rec = wfdb.rdrecord(os.path.join(data_path, record))
        ann = wfdb.rdann(os.path.join(data_path, record), 'apn')
        ecg = rec.p_signal[:, 0]
        fs = int(rec.fs)

        ecg = bandpass_filter(ecg, fs)
        ecg = normalize(ecg)

        samples_per_minute = fs * 60

        for i, start in enumerate(ann.sample):
            end = start + samples_per_minute
            if end > len(ecg):
                continue
            segment = ecg[start:end]
            label = ann.symbol[i]  # 'A' or 'N'

            rr_feats = extract_rr_features(segment, fs)
            rr_feats['label'] = label
            rr_feats['record'] = record
            rr_feats['minute_index'] = i
            features_data.append(rr_feats)

            print(f"{record} | Segment {i} | Label: {label} | Features: {rr_feats}")

    except Exception as e:
        print(f"Error in {record}: {e}")

# Save features to CSV
df = pd.DataFrame(features_data)
df.to_csv(output_csv, index=False)
print(f"\n✅ Features saved to: {output_csv}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import numpy as np

# Load features
df = pd.read_csv('/content/drive/MyDrive/apnea-ecg/rr_features.csv')

# Encode labels
df['label'] = LabelEncoder().fit_transform(df['label'])  # 'A'=0, 'N'=1

# Feature matrix and target
X = df[['rr_mean', 'rr_std', 'rr_min', 'rr_max', 'rr_count']]
y = df['label']

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader

# Move to GPU if available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Prepare tensors and dataloaders
X_train_cnn = torch.tensor(X_train, dtype=torch.float32).unsqueeze(1).to(device)  # (batch, 1, features)
X_test_cnn  = torch.tensor(X_test,  dtype=torch.float32).unsqueeze(1).to(device)
y_train_cnn = torch.tensor(y_train.values, dtype=torch.long).to(device)
y_test_cnn  = torch.tensor(y_test.values, dtype=torch.long).to(device)

train_loader = DataLoader(TensorDataset(X_train_cnn, y_train_cnn), batch_size=32, shuffle=True)
test_loader  = DataLoader(TensorDataset(X_test_cnn,  y_test_cnn),  batch_size=32)

# Model definition
class CNNClassifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv1d(1, 16, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv1d(16, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.AdaptiveAvgPool1d(1)
        )
        self.fc = nn.Linear(32, 2)

    def forward(self, x):
        x = self.conv(x)
        x = x.view(x.size(0), -1)
        return self.fc(x)

# Instantiate, move to device
model     = CNNClassifier().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training loop
for epoch in range(10):
    model.train()
    total_loss = 0.0
    for xb, yb in train_loader:
        optimizer.zero_grad()
        logits = model(xb)
        loss   = criterion(logits, yb)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1}/10  —  Train Loss: {total_loss/len(train_loader):.4f}")

# Evaluation on test set
model.eval()
correct = 0
total   = 0
with torch.no_grad():
    for xb, yb in test_loader:
        logits = model(xb)
        preds  = torch.argmax(logits, dim=1)
        correct += (preds == yb).sum().item()
        total   += yb.size(0)

accuracy = correct / total * 100
print(f"\n✅ CNN Test Accuracy: {accuracy:.2f}%")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader

# 1. Device setup
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# 2. Prepare tensors
# (Assuming X_train, X_test, y_train, y_test are already defined as numpy arrays/pandas series)
X_train_lstm = torch.tensor(X_train, dtype=torch.float32).unsqueeze(1).to(device)
X_test_lstm  = torch.tensor(X_test,  dtype=torch.float32).unsqueeze(1).to(device)
y_train_lstm = torch.tensor(y_train.values, dtype=torch.long).to(device)
y_test_lstm  = torch.tensor(y_test.values, dtype=torch.long).to(device)

train_loader = DataLoader(TensorDataset(X_train_lstm, y_train_lstm), batch_size=32, shuffle=True)
test_loader  = DataLoader(TensorDataset(X_test_lstm,  y_test_lstm),  batch_size=32)

# 3. Define your LSTM model
class LSTMClassifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.lstm = nn.LSTM(input_size=X_train.shape[1], hidden_size=64, batch_first=True)
        self.fc   = nn.Linear(64, 2)

    def forward(self, x):
        _, (h_n, _) = self.lstm(x)
        return self.fc(h_n.squeeze(0))

# 4. Instantiate, move to device, define loss & optimizer
model     = LSTMClassifier().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 5. Training loop
num_epochs = 10
for epoch in range(num_epochs):
    model.train()
    total_loss = 0.0
    for xb, yb in train_loader:
        optimizer.zero_grad()
        logits = model(xb)
        loss   = criterion(logits, yb)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1}/{num_epochs} — Train Loss: {total_loss/len(train_loader):.4f}")

# 6. Evaluation on test set
model.eval()
correct = 0
total   = 0
with torch.no_grad():
    for xb, yb in test_loader:
        logits = model(xb)
        preds  = torch.argmax(logits, dim=1)
        correct += (preds == yb).sum().item()
        total   += yb.size(0)

accuracy = correct / total * 100
print(f"\n✅ LSTM Test Accuracy: {accuracy:.2f}%")
```
